Correlation is a statistical measure that describes the extent to which two variables are dependent on each other. It ranges from -1 to 1, where -1 indicates a perfect negative correlation, 1 indicates a perfect positive correlation, and 0 indicates no correlation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.rand(100) * 10
y = 2 * x + 1 + np.random.randn(100) * 2
```

```python
# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(x, y)[0, 1]
```

```python
plt.scatter(x, y, label=f'Correlation: {correlation_coefficient:.2f}')
plt.title('Scatter Plot with Correlation')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```
